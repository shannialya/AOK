# Menghitung Nilai Korelasi
Menghitung nilai korelasi dengan metode Kendall dan Spearman

## Metode Kendall
Statistik yang digunakan untuk mengukur hubungan ordinal antara dua kuantitas yang diukur. Korelasi Kendall antara dua variabel akan tinggi ketika pengamatan memiliki peringkat yang sama dan rendah ketika pengamatan memiliki peringkat yang berbeda atau sepenuhnya berbeda untuk korelasi antara kedua variabel.

```r
correlations = datarespon.corr(method ="kendall") 
correlations
```

Kesimpulan: pada metode Kendall menunjukkan dua pertanyaan yang memiliki skor tertinggi diantara pertanyaan lainnya, yaitu:
1. Pekerjaan desainer grafis sangat diperlukan di masa sekarang
2. Pekerjaan dari desain grafis mendapat banyak keuntungan dari dalam perdagangan

Kedua pertanyaan ini mendapatkan nilai korelasi sebesar 0.317891 yang mengindikasikan kedua pertanyaan ini memiliki hubungan yang saling terkait. Kesimpulan yang dapat diambil adalah dikarenakan pekerjaan desainer grafis sangat dibutuhkan di masa sekarang (masa era teknologi modern) dimana industri-industri perdagangan sangat membutuhkan jasa desainer grafis, sehingga desainer grafis mendapatkan banyak sekali keuntungan di dalam dunia perdagangan.

## Metode Spearman
Korelasi Spearman tidak mengasumsikan bahwa data berasal dari distribusi tertentu, jadi itu adalah ukuran korelasi non-parametrik. Korelasi Spearman juga dikenal sebagai korelasi peringkat Spearman karena menghitung koefisien korelasi pada nilai peringkat data.

```r
correlations = datarespon.corr(method ="spearman") 
correlations
```

## Map Korelasi

```r
# correlation map
## higher correlations are brighter
f,ax = plt.subplots(figsize=(20,20))
sns.heatmap(correlations, annot=True, linewidths=.5, fmt='.3f', ax=ax)
plt.show
```

Score PCA memiliki korelasi antara satu dengan yang lainnya. Korelasi yang timbul dapat bersifat positif atau negatif. Korelasi score PCA dilihat dari koefisien korelasi masingâ€“masing pertanyaan satu dengan pertanyaan lain disebut dengan Score Correlation . Angka yang bercetak tebal menunjukkan korelasi antar atribut. Jika nilai korelasi suatu atribut dengan atribut lain bernilai 0.5, atribut tersebut dapat dikatakan berkorelasi, sedangkan jika nilai korelasinya lebih dari 0.8, atribut tersebut dapat dikatakan berkorelasi tinggi (Limpawattana, Shewfelt, 2010).
